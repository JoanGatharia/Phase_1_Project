# 3.Data Preparation

a. Data Cleaning
  - Missing Values
  - Duplicates

```python
#Check for any missing values
df.isnull().sum()
```

Observation 1
- I noticed that we have a lot of missing data in some of our columns and since the columns are needed for us to meet our objectives we will only drop the rows with missing values and not the entire column. We can use the subsets for that or replace Method

Missing Values
- Handling Missing values differs with the type of data you are working with
- we can either replace the missing data with measures of central tendency like Mean,Median and Mode, we could drop the entire column or we could select the specific rows that have missing values in a column and drop them
  1. Categorical data such as the Make, Model, Engine.Type,Aircraft.Category,Location,Country,Aircraft.damage,Injury.Severity we replace with the Mode
  2. To see which columns to replace missing values with the Median we will check for skewness.
  3. For replacement with the mean we need to be dealing with Normal distribution

```python
##For categorical data we replace the missing values using Mode
mode_values={'Make':'Unknown','Model':'Unknown','Engine.Type':'Unknown','Aircraft.Category':'Unknown','Location':'Unknown','Country':'Unknown','Aircraft.damage':'Unknown','Injury.Severity':'Unknown','Broad.phase.of.flight':'Unknown','Weather.Condition':'Unknown'} #Identified the columns we need the mode values
df=df.fillna(value=mode_values) #Use .fillna to fill the mode values on the missing value rows
df.isnull().sum() #Check if the columns still have missing values
```

- For the remaining columns that still have missing data and they are not categorical we need to check for skewness first To know if we replace with the mean, Median or we drop the row

```python
##Checking for Skewness
skewness=df[['Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured']].skew()
skewness
```

- I choose to check for skewness to see how to appropriately replace this numerical data columns.
- They are positively skewed (Right-Skewed) since the values are positive.
- This means we can replace using the Median/Mean
- From a glance we can also spot an outlier even before plotting.The value 87.269 is significantly higher than the other values
- I choose median because the data is skewed and not symmetric

```python
#Replacing using the median
median_values={'Total.Fatal.Injuries':df['Total.Fatal.Injuries'].median(),'Total.Serious.Injuries':df['Total.Serious.Injuries'].median(),'Total.Minor.Injuries':df['Total.Minor.Injuries'].median(),'Total.Uninjured':df['Total.Uninjured'].median()}
df=df.fillna(value=median_values)
df.isnull().sum() #Check if the columns still have missing values
```

Observation 2
- We have successfully replaced all the missing values now let's check the duplicates

```python
#Check for any duplicate values/Rows
df.duplicated().sum()
```

Observation 3
 - We have 2 duplicates that need to be dropped


```python
#Remove the duplicates
df=df.drop_duplicates()
```

```python
#We check if we still have duplicates
df.duplicated().sum()
```

A. Skewness Histogram

```python
#Let's start by confirming if indeed it was a positive skew
# Apply log transformation
df['Log.Fatal.Injuries'] = np.log1p(df['Total.Fatal.Injuries'])

plt.figure(figsize=(10, 6))
plt.hist(data=df, x='Log.Fatal.Injuries', bins=20, color='purple')
plt.title('Log-Transformed Histogram of Total Fatal Injuries')
plt.xlabel('Log of Total Fatal Injuries')
plt.ylabel('Frequency')
plt.show()
```

- The log transformation reduces skewness by compressing extreme values

- I also looked at the correlation between the Aircraft Model and Injury severity focusing on the Top 10 Models

```python
# Get the top 10 aircraft models by accident frequency
top_models = df['Model'].value_counts().nlargest(10).index

# Filter the data to include only the top models
filtered_df = df[df['Model'].isin(top_models)]

# Create the pivot table using the filtered data
heatmap_data = filtered_df.pivot_table(
    index='Model',
    columns='Injury.Severity',
    values='Accident.Number',
    aggfunc='count',
    fill_value=0
)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='Blues', fmt='d')  # Format annotations as integers
plt.title('Relationship between Aircraft Model and Injury Severity (Top 10 Models)')
plt.xlabel('Injury Severity')
plt.ylabel('Aircraft Model')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.yticks(rotation=0)  # Keep y-axis labels horizontal
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

- Air Craft Model 180 has the least number of fatal Injuries in comparison to the other Models
- Air Craft Model 150M has the Lowest number of Non-Fatal Injuries in comparison to other models having 515 non- fatal injuries
- Aircraft Model 152 has the highest number of non-fatal injuries having 2093
- Also noticed that Aircraft Model 152 has the highest number of Fatal Injuries
- In conclusion purchasing Aircraft Model 152 would be more risky in comparison to the other models

C.Bar Plots
- Let's Look at Aircraft Models with the Highest and the Lowest Accident rates as well as by the Accident count
- we will have to create a new column to work with called Accident_rates since we do not have such a column

```python
# Group by Model and count accidents
accident_counts = df.groupby('Model')['Accident.Number'].count()
   # Calculate total accidents
total_accidents = accident_counts.sum()
   # Calculate accident rates per model
accident_rates = (accident_counts / total_accidents) * 100
accident_rates
```

```python
#Create a new column called Accident_rates
df['Accident_rates']=df['Model'].map(accident_rates)
df
```

```python
#Air craft models with highest accident rates
#Sort by Accident_Rates and select the top 10
top_models = df.sort_values(by=['Accident_rates'], ascending=False).head(10)
top_models
```

```python
#Lets look at the model accident counts
Model_accident_counts = df.groupby(['Model']).size().reset_index(name='Accident_Count')
Model_accident_counts = Model_accident_counts.sort_values(by='Accident_Count', ascending=False)
Model_accident_counts
```

```python
#Lets plot the Aircraft Models with the Highest accident rates
#First i have to group the Aircraft models that are similar
grouped_models = df.groupby('Model').agg({
    'Accident_rates': 'mean',
    'Accident.Number': 'count'
}).reset_index() #This ensures that duplicate entries for the same model are combined
#Then sort by accident rates and select the top 10 with the highest Accident rates
top_models_by_rate = grouped_models.sort_values(by='Accident_rates', ascending=False).head(10)
#Then now we plot
plt.figure(figsize=(10, 6)) 
#to select darker colors
colors=['#2C3E50', '#3498DB', '#9B59B6', '#E67E22', '#2ECC71', '#F1C40F', '#7F7F7F', '#D35400', '#FF8C00', '#FFD700']
plt.barh(top_models_by_rate['Model'], top_models_by_rate['Accident_rates'],color=colors)
plt.xlabel('Aircraft Model')
plt.ylabel('Accident Rates')
plt.title('Top 10 Aircraft Models by Accident Rates')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

- When we use accident rates we still notice that Aircraft Model 152 has the highest number

```python
#Create a new column Accident count since we had not added it as a new column
df['Accident_Count']=df['Model'].map(Model_accident_counts.set_index('Model')['Accident_Count'])
df
```

```python
#Aircraft Models relationship with Accident counts
plt.figure(figsize=(10, 6))

top_models_by_accident = Model_accident_counts.head(10)
colors=['green','blue','purple','red','orange','yellow','pink','brown','gray','black']

plt.bar(top_models_by_accident['Model'], top_models_by_accident['Accident_Count'],color=colors)
plt.xlabel('Aircraft Model')
plt.ylabel('Accident Count')
plt.title('Top 10 Aircraft Models by Accident Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
#I decided to also look at the common types of aircraft damage across models through a bar plot
plt.figure(figsize=(10, 6))  # Increase figure size
damage_counts = df['Aircraft.damage'].value_counts().head(10)
colors=['green','blue','purple','red','orange','yellow','pink','brown','gray','black']
plt.bar(damage_counts.index, damage_counts.values,color=colors)
plt.xlabel('Aircraft Damage')
plt.ylabel('Count')
plt.title('Common Types of Aircraft Damage')

plt.show()
```

```python
# Common Types of Aircraft damage by Model for top 10 models

# We group by Aircraft Model and Damage Type and Damage count
grouped_damage = df.groupby(['Model', 'Aircraft.damage']).size().reset_index(name='Damage_Count')

# We sort by accident forcusing on top 10 models with the most accidents
top_10_models= grouped_damage.groupby('Model').sum().sort_values(by='Damage_Count', ascending=False).head(10).reset_index()
filtered_top_10_models = top_10_models[top_10_models['Model'].isin(top_models_by_rate['Model'])]

colors=['purple','green','blue','red','brown','gray','black']

# Common Types of Aircraft damage by Model for top 10 models

# We group by Aircraft Model and Damage Type and Damage count
grouped_damage = df.groupby(['Model', 'Aircraft.damage']).size().reset_index(name='Damage_Count')

#Pivot the data to get aircraft damage as the columns
damage_pivot = grouped_damage.pivot(index='Model', columns='Aircraft.damage', values='Damage_Count').fillna(0)

# We sort by accident forcusing on top 10 models with the most accidents
top_10_models=damage_pivot.sum(axis=1).sort_values(ascending=False).head(10).index

# Filter the data to include only the top 10 models
filtered_data = damage_pivot.loc[top_10_models]

colors=['purple','green','blue','red','brown','gray','black']

#We now plot
ax=filtered_data.plot(kind='bar', stacked=True, figsize=(10, 6),color=colors)
plt.xlabel('Aircraft Model')
plt.ylabel('Damage Count')
plt.title('Common Types of Aircraft Damage by Model')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

- Again we still see Aircraft Model 152 has more damage with comparison to the rest
- Aircraft model 180 seems to be much better in terms of experiencing damage since it has the lowest number of destroyed aircrafts

```python
#We can plot Aircraft models by damage count and check the relationship
top_10_models= grouped_damage.groupby('Model').sum().sort_values(by='Damage_Count', ascending=False).head(10).reset_index()
filtered_data = top_10_models[top_10_models['Model'].isin(top_models_by_rate['Model'])]


plt.figure(figsize=(12, 6))  # Resize plot
colors=['purple','green','blue','red','brown','gray','black']
plt.barh(filtered_data['Model'], filtered_data['Damage_Count'],color=colors)

plt.ylabel('Aircraft Model')
plt.xlabel('Count of Damage')
plt.title('Top 10 Aircraft Models by Damage Count')
plt.tight_layout()
plt.show()
```

```python
# I decided to look at the Engine types with the lowest accident count
engine_accidents = df.groupby('Engine.Type')['Accident.Number'].count().reset_index() # Group by engine type and count accidents

engine_accidents = engine_accidents.sort_values(by=['Accident.Number'], ascending=True) # Sort by accident count in ascending order (to find lowest counts)

# Select the top 6 engine types with the lowest accident counts
top_n = 6 # you can change this value to select a different number of engine types
lowest_accident_engines = engine_accidents.head(top_n)

colors=['brown','green','blue','purple','red','orange','yellow','pink','gray','black']

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(lowest_accident_engines['Engine.Type'], lowest_accident_engines['Accident.Number'], color=colors)
plt.xlabel('Engine Type')
plt.ylabel('Accident Count')
plt.title(f'Top {top_n} Engine Types with Lowest Accident Counts')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

D. Count Plot

```python
#We can plot the relationship between Aircraft model and engine type
#Lets work with like 10 models
top_10_models = df.groupby('Model').size().nlargest(10).index
filtered_df = df[df['Model'].isin(top_10_models)]

plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_df, x='Model', hue='Engine.Type', palette='viridis') #We use a count plot because we are plotting against the count of occurences
plt.xlabel('Aircraft Model')
plt.ylabel('Count')
plt.title('Relationship between Aircraft Model and Engine Type')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.tight_layout()
plt.show()
```

- We can see that the Reciprocating engine type is more prone to accidents since model 152 have more accident occurrences that were fatal in comparison to other models

E. Line graph
- I thought of creating a line graph showing the accident rates through the years to help determine if old aircraft models were better than the new Aircraft models

```python
#Change the event.date column to year
df['Event.Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Event.Date'].dt.year

#Plot a line graph showing accident rate progression through the years
plt.figure(figsize=(10, 6))
yearly_accident_rate = df.groupby('Year').size()
yearly_accident_rate.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Accident Count')
plt.title('Progression of Accident Count Through the Years')
plt.grid(True)
plt.show()
```

- From the above Line graph we are able to notice that aircrafts built from the year 1950- 1980 had almost no accident occurences however between 1982-2000 the accident rates were higher but then significantly dropped afterwards
- This observation means that old aircraft models had a better make thus less prone to accidents meaning purchasing recent models or aircrafts with new model engines would be more safe and economically advantagious to the company

F. Pie Chart
- I choose a pie chart for the relationship between the types of injuries just to clearly see by percentage which one was higher

```python
#A pie chart to show relationship of the injuries by percentage
# Calculate the total number of injuries
total_injuries = df['Total.Fatal.Injuries'].sum() + df['Total.Serious.Injuries'].sum() + df['Total.Minor.Injuries'].sum() + df['Total.Uninjured'].sum()
#A pie chart
labels = ['Total Fatal Injuries', 'Total Serious Injuries', 'Total Minor Injuries', 'Total Uninjured']
colors = ['red', 'green', 'yellow', 'orange']
plt.figure(figsize=(8, 8))
plt.pie([df['Total.Fatal.Injuries'].sum(), df['Total.Serious.Injuries'].sum(), df['Total.Minor.Injuries'].sum(), df['Total.Uninjured'].sum()], labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Relationship of Injuries by Percentage')
plt.axis('equal')
plt.show()
```
